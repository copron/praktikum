```python
# функции из moduls.py
from moduls import *

def main():
    while True:
        print("\nФайловый менеджер")
        print("1. Создать директорию")
        print("2. Удалить директорию")
        print("3. Показать содержимое текущей директории")
        print("4. Изменить текущую директорию")
        print("5. Создать файл")
        print("6. Прочитать содержимое файла")
        print("7. Записать содержимое в файл")
        print("8. Удалить файл")
        print("9. Скопировать файл")
        print("10. Переместить файл")
        print("11. Переименовать файл")
        print("12. Выход")

        choice = input("Выберите действие: ")

        if choice == '1':
            directory_name = input("Введите имя директории: ")
            create_directory(directory_name)
        elif choice == '2':
            directory_name = input("Введите имя директории: ")
            delete_directory(directory_name)
        elif choice == '3':
            list_contents()
        elif choice == '4':
            new_directory = input("Введите новый путь к директории: ")
            change_directory(new_directory)
        elif choice == '5':
            file_name = input("Введите имя файла: ")
            content = input("Введите содержимое файла (если есть): ")
            create_file(file_name, content)
        elif choice == '6':
            file_name = input("Введите имя файла: ")
            read_file(file_name)
        elif choice == '7':
            file_name = input("Введите имя файла: ")
            content = input("Введите новое содержимое файла: ")
            write_file(file_name, content)
        elif choice == '8':
            file_name = input("Введите имя файла: ")
            delete_file(file_name)
        elif choice == '9':
            source_file = input("Введите имя исходного файла: ")
            destination = input("Введите директорию: ")
            copy_file(source_file, destination)
        elif choice == '10':
            source_file = input("Введите имя исходного файла: ")
            destination = input("Введите директорию: ")
            move_file(source_file, destination)
        elif choice == '11':
            old_name = input("Введите старое имя файла: ")
            new_name = input("Введите новое имя файла: ")
            rename_file(old_name, new_name)
        elif choice == '12':
            print("Выход из программы.")
            break
        else:
            print("Некорректный выбор. Попробуйте еще раз.")

if __name__ == "__main__":
    main()
```
